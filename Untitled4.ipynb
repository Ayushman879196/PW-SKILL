Question 1: What is Anomaly Detection? Explain its types (point, contextual, and collective anomalies) with examples.


Anomaly Detection is the process of identifying rare items, events, or observations which deviate significantly from the majority of the data and do not conform to an expected pattern. Anomalies are also referred to as outliers, novelties, noise, deviations, and exceptions. Here are its types with examples:

Point Anomalies: A single data instance is anomalous if it deviates significantly from the rest of the data. This is the simplest type of anomaly.

Example: A sudden, unusually high credit card transaction amount for a user who typically makes small purchases.

Contextual Anomalies: A data instance is anomalous in a specific context, but not otherwise. The context could be temporal, spatial, or behavioral.

Example: A temperature reading of 30°C in a desert in summer is normal, but the same temperature in Antarctica during winter would be a contextual anomaly.

Collective Anomalies: A collection of related data instances is anomalous with respect to the entire dataset, even if individual data instances within the collection are not themselves anomalous. The anomaly is in the sequence or structure of the data.

Example: A server that sends a series of identical small data packets over a short period might not be considered anomalous individually, but the collective pattern of these packets could indicate a DDoS attack.

Question 2: Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of their approach and suitable use cases.


1. Isolation Forest
Approach: Isolation Forest is an unsupervised anomaly detection algorithm that works on the principle of isolation. It isolates anomalies instead of profiling normal data points. It builds an ensemble of isolation trees (similar to random forests). Anomalies are points that are easier to isolate (i.e., require fewer splits to be separated from the rest of the data in a tree).
How it works: It randomly selects a feature and then a random split value between the maximum and minimum values of the selected feature. This partitioning is repeated recursively until each instance is isolated or a maximum tree depth is reached. Anomalies are isolated closer to the root of the tree (fewer splits), while normal points are isolated deeper in the tree (more splits).
Suitable Use Cases:
High-dimensional datasets: It performs well in datasets with many features because it doesn't need to compute distance metrics.
Large datasets: It has a linear time complexity, making it efficient for large datasets.
Numerical data: Primarily designed for numerical data.
Detecting global anomalies: Excellent for identifying points that are distinctly different from the majority of the data.
Unlabeled data: Effective when there's no prior knowledge about anomalies.
2. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)
Approach: DBSCAN is a density-based clustering algorithm that can identify anomalies as noise points. It groups together data points that are closely packed together (points with many nearby neighbors), marking as outliers those points that lie alone in low-density regions.
How it works: It requires two parameters: eps (the maximum distance between two samples for one to be considered as in the neighborhood of the other) and minPts (the number of samples in a neighborhood for a point to be considered as a core point). Points are classified as core points, border points, or noise points (anomalies).
Suitable Use Cases:
Irregularly shaped clusters: Can discover clusters of arbitrary shape, unlike K-means.
Spatial data: Particularly useful for detecting anomalies in spatial datasets.
Detecting local anomalies: Good at finding anomalies that are anomalous within their local neighborhood.
Varying density clusters: Can handle clusters of varying densities.
When the number of clusters is unknown: It does not require the number of clusters to be specified beforehand.
3. Local Outlier Factor (LOF)
Approach: LOF is a density-based outlier detection algorithm that measures the local deviation of density of a given data point with respect to its neighbors. It compares the local density of a point to the local densities of its neighbors. If a point has a significantly lower local density than its neighbors, it's considered an outlier.
How it works: It calculates the reachability distance and local reachability density for each point. An LOF score is then assigned to each data point, representing its degree of outlierness. A score close to 1 indicates a normal point, while significantly higher values indicate anomalies.
Suitable Use Cases:
Detecting local anomalies: Extremely effective at identifying anomalies that are abnormal relative to their immediate surroundings, even if they are not globally extreme.
Heterogeneous data densities: Performs well in datasets where the density of normal points can vary across different regions.
When context matters: Suitable when the notion of an outlier depends on the local distribution of data.
Small to medium-sized datasets: Computationally more intensive than Isolation Forest for very large datasets.
Numerical data: Best suited for numerical data where distance metrics are meaningful.

Question 3: What are the key components of a Time Series? Explain each with one example.

Here's an explanation of each with an example:

1. Trend
Explanation: The trend component refers to the long-term increase or decrease in the data. It reflects the underlying direction of the series over an extended period, ignoring short-term fluctuations.
Example: The general increase in global average temperatures over the past century is a long-term upward trend, indicating climate change.
2. Seasonality
Explanation: Seasonality refers to patterns that repeat over a fixed period, such as a day, week, month, or year. These patterns are typically caused by factors like weather, holidays, or recurring social/business events.
Example: The sales of ice cream typically increase every summer and decrease every winter, showing a clear seasonal pattern on a yearly basis.
3. Cyclicality
Explanation: Cyclicality describes patterns that oscillate around the trend, but unlike seasonality, they do not have a fixed period. The duration of a cycle is usually longer than a year and can be influenced by economic conditions (e.g., business cycles) or other long-term phenomena.
Example: Periods of economic expansion followed by recession, which can last several years, represent a cyclical pattern in economic indicators like GDP or unemployment rates.
4. Irregularity (Noise or Residual)
Explanation: The irregular component, also known as noise or residual, represents the unpredictable, random fluctuations in the time series that cannot be explained by trend, seasonality, or cyclicality. These are often random disturbances or unpredictable events.
Example: A sudden, unexpected surge in electricity consumption due to an unforecasted heatwave is an irregular component, as it's a one-off event not part of a recurring pattern or long-term trend.

Question 4: Define Stationary in time series. How can you test and transform a non-stationary series into a stationary one?

What is Stationarity in Time Series?
A time series is said to be stationary if its statistical properties, such as mean, variance, and autocorrelation, are constant over time. This means that the way the series behaves at one point in time is the same as how it behaves at any other point in time. Stationarity is a fundamental assumption for many time series models (e.g., ARIMA models), as non-stationary series can lead to spurious correlations and unreliable forecasts.

There are generally two types of stationarity:

Strict Stationarity: The joint distribution of any set of observations remains the same regardless of where they are observed in time. This is a very strong condition and rarely met in real-world data.
Weak-form (or Covariance) Stationarity: A less restrictive form, which requires:
Constant Mean: The mean of the series does not depend on time.
Constant Variance: The variance of the series does not depend on time.
Constant Autocorrelation: The covariance between two values depends only on the lag between them, not on the actual time at which the values are observed.
How to Test for Stationarity?
You can test for stationarity using both visual inspection and statistical tests.

1. Visual Inspection:
Plotting the Time Series: Look for trends (increasing or decreasing mean), seasonality (recurring patterns), and changes in variability (increasing or decreasing variance) over time. If any of these are present, the series is likely non-stationary.
Rolling Statistics: Plotting rolling mean and rolling standard deviation over time can visually confirm if these statistics are constant. If they fluctuate significantly, the series is non-stationary.
Autocorrelation Function (ACF) and Partial Autocorrelation Function (PACF) Plots: For a stationary series, the ACF plot will drop off rapidly to zero, while for a non-stationary series (especially with a trend), it will typically decay slowly.
2. Statistical Tests:
Augmented Dickey-Fuller (ADF) Test: This is one of the most common statistical tests for stationarity. It tests the null hypothesis that a unit root is present in the time series (i.e., the series is non-stationary). If the p-value is less than a chosen significance level (e.g., 0.05), you can reject the null hypothesis and conclude that the series is stationary.
Interpretation: A low p-value (e.g., < 0.05) and a test statistic more negative than the critical values suggest stationarity.
Kwiatkowski-Phillips-Schmidt-Shin (KPSS) Test: This test is an alternative to the ADF test. Its null hypothesis is that the time series is stationary (i.e., no unit root). Therefore, if the p-value is less than the significance level, you reject the null hypothesis and conclude that the series is non-stationary.
Interpretation: A high p-value (e.g., > 0.05) suggests stationarity.
How to Transform a Non-Stationary Series into a Stationary One?
If a time series is found to be non-stationary, various transformations can be applied to make it stationary. The goal is to remove trends, seasonality, and heteroscedasticity (non-constant variance).

Differencing: This is the most common method to remove trends and seasonality. It involves subtracting the observation at the previous time step from the current observation.

First-order differencing: Y_t = X_t - X_{t-1} (removes linear trend).
Seasonal differencing: Y_t = X_t - X_{t-L} (removes seasonality of length L).
Example: If your data has a linear trend, taking the first difference will often stabilize the mean.
Transformation (e.g., Log Transformation): This is primarily used to stabilize the variance (i.e., address heteroscedasticity) and can also help in making trends more linear.

Example: If the variance of your series increases with the mean, applying a logarithm (e.g., log(X_t)) can compress the larger values and stabilize the variance.
Decomposition: Separating the time series into its trend, seasonal, and residual components. Once separated, you can model the stationary residual component.

Moving Average (Smoothing): While not a direct transformation to achieve stationarity, smoothing techniques can help in identifying underlying trends or patterns that can then be differenced.


Question 5: Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in terms of structure and application.

1. Autoregressive (AR) Model
Structure: An AR model of order p (AR(p)) uses a linear combination of its own past observations to predict future values. It assumes that the current value of the time series depends on its previous values.
Equation: Y_t = c + φ_1*Y_{t-1} + φ_2*Y_{t-2} + ... + φ_p*Y_{t-p} + ε_t
Y_t: Current value
c: Constant
φ_i: Autoregressive coefficients
Y_{t-i}: Past values
ε_t: White noise error term
Application: Best suited for time series where the current value is a linear function of its past values. Often used for modeling phenomena with inertia, where past behavior directly influences future behavior (e.g., stock prices, temperature).
2. Moving Average (MA) Model
Structure: An MA model of order q (MA(q)) uses a linear combination of past forecast errors (or white noise error terms) to predict future values. It assumes that the current value depends on past random shocks.
Equation: Y_t = c + ε_t + θ_1*ε_{t-1} + θ_2*ε_{t-2} + ... + θ_q*ε_{t-q}
Y_t: Current value
c: Constant
θ_i: Moving average coefficients
ε_{t-i}: Past error terms
ε_t: White noise error term
Application: Useful for time series where fluctuations are mainly driven by random shocks or unforeseen events that have a decaying effect over time (e.g., modeling financial market volatility due to sudden news).
3. Autoregressive Integrated Moving Average (ARIMA) Model
Structure: ARIMA(p, d, q) combines the AR and MA models with an 'Integrated' component (I). The I stands for differencing, which is used to make a non-stationary time series stationary. d is the number of times differencing is applied.
Equation: The differenced series Z_t = (1-L)^d Y_t is modeled as an ARMA(p, q) process.
p: Order of the AR part.
d: Number of non-seasonal differences needed for stationarity.
q: Order of the MA part.
Application: Widely used for forecasting non-seasonal time series that exhibit trends and can be made stationary by differencing. Applicable in many fields, including economics, finance, and engineering, for forecasting sales, economic indicators, etc.
4. Seasonal Autoregressive Integrated Moving Average (SARIMA) Model
Structure: SARIMA(p, d, q)(P, D, Q)s extends ARIMA to handle time series with both non-seasonal and seasonal components. It incorporates seasonal AR, seasonal MA, and seasonal differencing.
(p, d, q): Non-seasonal AR, differencing, and MA orders.
(P, D, Q): Seasonal AR, differencing, and MA orders.
s: The length of the seasonal period (e.g., 12 for monthly data, 4 for quarterly data).
Application: Essential for time series that display clear seasonal patterns in addition to trends. Common in sales forecasting, energy consumption prediction, and any data influenced by calendar cycles (e.g., monthly retail sales, quarterly utility usage).
5. Seasonal Autoregressive Integrated Moving Average with Exogenous Regressors (SARIMAX) Model
Structure: SARIMAX(p, d, q)(P, D, Q)s adds the capability to include exogenous (external) variables (X) to the SARIMA model. These external variables are not part of the time series itself but can influence its behavior.
Equation: Similar to SARIMA, but includes an additional term β_k*X_{t,k} for each exogenous variable X_k.
Application: Used when external factors significantly impact the time series and can be measured. For instance, forecasting ice cream sales might use SARIMAX to include temperature or holiday indicators as exogenous variables, alongside the historical sales data and seasonal patterns. This allows for a more comprehensive model by incorporating external influences.

```python
#Question 6: Time Series Decomposition (AirPassengers)
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Load the AirPassengers dataset using get_rdataset
# The 'datasets' argument refers to the R datasets package which includes AirPassengers
air_passengers_r = sm.datasets.get_rdataset("AirPassengers", "datasets")
data_df = air_passengers_r.data

# The 'value' column contains the passenger numbers
# The 'time' column represents fractional years (e.g., 1949.0 for Jan 1949, 1949.0833 for Feb 1949)
# We need to convert this to a proper DatetimeIndex

# AirPassengers data runs from 1949-01 to 1960-12, 144 observations.
start_date = pd.to_datetime('1949-01-01')
num_months = len(data_df)
date_range = pd.date_range(start=start_date, periods=num_months, freq='MS')

# Create a new DataFrame with the correct DatetimeIndex and 'Passengers' column
data = pd.DataFrame(data_df['value'].values, index=date_range, columns=['Passengers'])
data.index.freq = 'MS' # Set the frequency to Monthly Start

# Display first few rows and info for verification
print("First 5 rows of the dataset:")
display(data.head())
print("\nDataset Information:")
data.info()

# Perform seasonal decomposition
decomposition = seasonal_decompose(data['Passengers'], model='multiplicative')

# Plot the decomposed components
decomposition.plot()
plt.show()
```

```python
#Question 7: Isolation Forest for Anomaly Detection
from sklearn.ensemble import IsolationForest
import pandas as pd
import matplotlib.pyplot as plt

# Load the nyc_taxi dataset from a public URL
df = pd.read_csv('https://raw.githubusercontent.com/numenta/NAB/master/data/realKnownCause/nyc_taxi.csv')

# Convert timestamp to datetime and set as index
df['timestamp'] = pd.to_datetime(df['timestamp'])
df = df.set_index('timestamp')

model = IsolationForest(contamination=0.01)
df['anomaly'] = model.fit_predict(df[['value']]) # Assuming 'value' is the column for anomaly detection in this dataset

# Plotting the anomalies
plt.figure(figsize=(15, 6))
plt.scatter(df.index, df['value'], c=df['anomaly'], cmap='coolwarm', s=20)
plt.title('Anomaly Detection using Isolation Forest on NYC Taxi Data')
plt.xlabel('Timestamp')
plt.ylabel('Value')
plt.show()
```

```python
#Question 8: SARIMA Forecasting
from statsmodels.tsa.statespace.sarimax import SARIMAX
model = SARIMAX(data['Passengers'], order=(1,1,1), seasonal_order=(1,1,1,12))
results = model.fit()
forecast = results.forecast(12)
print(forecast)
```
